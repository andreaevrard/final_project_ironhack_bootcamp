#### Dropping columns that don't present relevant information for predicting cost

```python
data = data.drop(columns=['food_category', 'food_department', 'food_family', 'gender', 'marital_status',
                         'brand_name', 'avg. yearly_income', 'education', 'member_card', 'houseowner', 
                          'sales_country', 'occupation'])
data.head()
#drop columns because there is very little change between unique values regarding cost, so they're not relevant features for predicting cost
```

#### Standardize numerical variables

```python
from sklearn.preprocessing import StandardScaler
for c in numericals:   # we'll normalize all column except the last one (the target column)
    data[c]=StandardScaler().fit_transform(data[c].values.reshape(-1, 1))
```

#### Encoding categorical variables

```python
categoricals = []
for column in data.columns:
    if (data[column].dtype == 'O'):
        categoricals.append(column)
```

```python
one_hot_data=pd.get_dummies(data[categoricals])   # one hot encoding categorical variables
one_hot_data.head()
```

#### Final dataset before modeling

```python
data=pd.concat([data, one_hot_data], axis=1)   # concat dataframes
data.drop(columns=categoricals, inplace=True)
data.head()
```

### Model

#### Separate variable between X and y (target) and train and test split

```python
X= data.drop(columns=['cost'])
y= data['cost']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=42)
```

#### Modeling

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression as LinReg
linreg=LinReg()    # model
linreg.fit(X_train, y_train)   # model train
y_pred_linreg=linreg.predict(X_test)   # model prediction
```

#### Regulation

```python
from sklearn.linear_model import Lasso     
from sklearn.linear_model import Ridge       
from sklearn.linear_model import ElasticNet
```

```python
# Lasso L1

lasso=Lasso()
lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
```

```python
# Ridge L2

ridge=Ridge()
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)
```

```python
# ElasticNet L1+L2

elastic=ElasticNet()
elastic.fit(X_train, y_train)

y_pred_elastic = elastic.predict(X_test)
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor as RFR

rfr=RFR()
rfr.fit(X_train, y_train)

y_pred_rfr = rfr.predict(X_test)
```

#### XGBoost

```python
# conda install -c conda-forge xgboost

from xgboost import XGBRegressor as XGBR

xgbr=XGBR()
xgbr.fit(X_train, y_train)

y_pred_xgbr = xgbr.predict(X_test)
```

#### LightGBM

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
iris = load_iris()
logistic = LogisticRegression(solver='saga', tol=1e-2, max_iter=200,
                                  random_state=0)
distributions = dict(C=uniform(loc=0, scale=600),
                     penalty=['l2', 'l1'])
clf = RandomizedSearchCV(logistic, distributions, random_state=0)
search = clf.fit(iris.data, iris.target)
search.best_params_
```

### Model Validation

```python
models=[linreg, lasso, ridge, elastic, rfr, xgbr]
model_names=['linreg', 'lasso', 'ridge', 'elastic', 'rfr', 'xgbr']
preds=[y_pred_linreg, y_pred_lasso, y_pred_ridge, y_pred_elastic, y_pred_rfr, y_pred_xgbr]
```

### R2

```python
for i in range(len(models)):

    train_score=models[i].score(X_train, y_train) #R2
    test_score=models[i].score(X_test, y_test)

    print ('Model: {}, train R2: {} -- test R2: {}'.format(model_names[i], train_score, test_score))
```

### MSE

```python
from sklearn.metrics import mean_squared_error as mse

for i in range(len(models)):

    train_mse=mse(models[i].predict(X_train), y_train) #MSE
    test_mse=mse(preds[i], y_test)

    print ('Model: {}, train MSE: {} -- test MSE: {}'.format(model_names[i], train_mse, test_mse))
```

### RMSE

```python
for i in range(len(models)):

    train_rmse=mse(models[i].predict(X_train), y_train)**0.5 #RMSE
    test_rmse=mse(preds[i], y_test)**0.5

    print ('Model: {}, train RMSE: {} -- test RMSE: {}'.format(model_names[i], train_rmse, test_rmse))
```

### MAE

```python
from sklearn.metrics import mean_absolute_error as mae
for i in range(len(models)):
    train_mae=mae(models[i].predict(X_train), y_train) #MAE
    test_mae=mae(preds[i], y_test)

    print ('Model: {}, train MAE: {} -- test MAE: {}'.format(model_names[i], train_mae, test_mae))
```
